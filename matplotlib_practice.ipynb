```python
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create plot 
plt.plot(x_axis, y_axis)
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')
#Create label for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit between 0 and 45. 
plt.ylim(0,45)
#Create a title
plt.title("PyBer Fare by Month")
#Add a legend
plt.legend()
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="Blue", linewidth=2, label = 'Boston')
#Create label for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set y limit between 0 and 45. 
plt.ylim(0,45)
#Create a title
plt.title("PyBer Fare by Month")
#Add a grid
plt.grid()
#Add a legend
plt.legend()
```

```python
#Create the plot and add a label for the legend
fig, ax = plt.subplots()

#Create label for the x and y axes
ax.plot(x_axis,y_axis,marker="D",color="Green", linewidth=2,label="Boston")
#Set y limit between 0 and 45. 
ax.set_ylim(0,45)
ax.set_xlim("Jan", "Dec")
#Create a title
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#Add a grid
ax.grid()
#Add a legend
ax.legend()
```

```python
# Create the bar plot 
plt.bar(x_axis, y_axis)
```

```python
#Create the plot. 
plt.bar(x_axis, y_axis, color="green", label='Boston')
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create Title
plt.title("PyBer Fare by Month")
#Add Legend
plt.legend()
```

```python
#Create the plot for horizontal 
plt.barh(x_axis,y_axis,color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.title("PyBer Fare by Month")

```

```python
#refactoring using loops within plt.scatter
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])
```

```python
#OOP Method
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*3 for i in y_axis], edgecolor='black')
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```

```python
#Creating pie charts
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
#OOP pie charts
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis, autopct='%.1f%%')
ax.invert_xaxis()
plt.show()
```
